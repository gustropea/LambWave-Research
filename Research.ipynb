# Fdisk


The earth geoid satisfies
\begin{equation}
\begin{aligned}
\frac{(x_p)^{2}}{a^{2}}+\frac{(y_p)^{2}}{b^{2}}+\frac{(z_p)^{2}}{c^{2}}&=1
\end{aligned}
\end{equation}

for $a, b,c$ ($a == b$?). 

The distance $d$ of a physical point $(x_p,y_p,z_p)$ in space to a smaller sphere with center $(x_c,y_c,z_c)$ is given by 
\begin{equation}
\begin{aligned}
d^2 = (x_p-x_c)^{2}+(y_p-y_c)^{2} + (z_p-z_c)^{2}
\end{aligned}
\end{equation}

The set of points on the earth geoid that lies inside the smaller sphere satisfies

\begin{equation}
S = \{(x_p,y_p,z_p) : d^2 \le R^2, (x_p,y_p,z_p) \quad \mbox{on the geoid} \}
\end{equation}

So that we can define `fdisc` as 

       d2 = (xp-xc)**2 + (yp-yc)**2 + (zp-zc)**2
       fdisc = d**2  - R**2

After combining the two equations:
    
\begin{equation}
\begin{aligned}
    \frac{2(xc)(xp)}{a^{2}}+\frac{2(yc)(yp)}{b^{2}}+\frac{2(zc)(zp)}{c^{2}} = \left(1- r^{2}\right)+\left(\frac{(xc)^{2}}{a^{2}}+\frac{(yc)^{2}}{b^{2}}+\frac{(zc)^{2}}{c^{2}}\right)
\end{aligned}
\end{equation}

There is no intersection if
$$\left(\frac{(xc)^{2}}{a^{2}}+\frac{(yc)^{2}}{b^{2}}+\frac{(zc)^{2}}{c^{2}}\right)>\left(1+r\right)^{2}$$
